```python
    import numpy as np
    import matplotlib.pyplot as plt
    from matplotlib import *
    import math as m
    import cmath
    import time
    from matplotlib import cm
    import skrf as rf
    from scipy.optimize import curve_fit
```

```python

omega_a=8.25e9
omega_ms = np.linspace(8.2,8.3,10001)*1e9
kaint=2.58e6
kaext=10e6
ka=kaint+kaext
gamma = 5.1e6
g_ma = 11e6*1j


omega1=(omega_a+omega_ms)/2+np.sqrt((omega_a-omega_ms)**2+g_ma**2)/2
omega2=(omega_a+omega_ms)/2-np.sqrt((omega_a-omega_ms)**2+g_ma**2)/2
    
plt.figure(figsize=(7, 7))
plt.plot(omega_ms/1e9,omega1/1e9,label=r'$\omega_+$',color='blue',linewidth=5)
plt.plot(omega_ms/1e9,omega2/1e9,label=r'$\omega_+$',color='red',linewidth=5)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

```python
omega_a=8.25e9
omega_ms=np.linspace(8.23,8.27,401)*1e9
omega_p=np.linspace(8.23,8.27,401)*1e9
# omega_s=np.linspace(8.3,8.32,2001)*1e9
#0.2/20001
#0.02/200
k_int=7e6
k_1=15.5e6
k_c=k_int+k_1

gamma_int=11.7e6

gamma=gamma_int

g=5e6
S=[]

for i in range(len(omega_ms)):
    R = 1 - k_1 / (-1j * (omega_p  - omega_a) + k_c / 2 + g ** 2*np.exp(1j*1*np.pi) / (
            -1j * (omega_p - omega_ms[i]) + gamma / 2))
    S.append(rf.mag_2_db(np.abs(R)))


plt.figure(figsize=(8, 4))
ax1 = plt.subplot(121)
gci1 = ax1.pcolor(omega_ms/1e9,omega_p/1e9,np.transpose(S))
cbar = plt.colorbar(gci1)
plt.show()
```
